```python
# Document Structure

from langchain_core.documents import Document
```

```python
doc = Document(
    page_content = "This is an example text content we are using for building a RAG",
    metadata={
        "source":"example.txt",
        "pages":1,
        "author":"Rinku Yadav",
        "date created":"2026-04-29"
    }
)
doc
```

```python
# Text Loader
from langchain_community.document_loaders import TextLoader
```

```python
txtloader = TextLoader("../data/python.txt",encoding = "utf-8")

txtDoc = txtloader.load()
print(txtDoc)
```

```python
#Directory Loader

from langchain_community.document_loaders import DirectoryLoader

dirLoader = DirectoryLoader(
    "../data",
    glob = "**/*.txt",
    loader_cls=TextLoader,
    loader_kwargs={"encoding":"utf-8"},
    show_progress=False
)

dirDoc = dirLoader.load()
print(dirDoc)
```

```python
# PDF loader

from langchain_community.document_loaders import PyMuPDFLoader

dirLoader = DirectoryLoader(
    "../data",
    glob = "**/*.pdf",
    loader_cls=PyMuPDFLoader,
    # loader_kwargs={"encoding":"utf-8"},
    show_progress=False
)

dirDoc = dirLoader.load()
print(dirDoc)
```
